```python
import pandas as pd 
import numpy as np #Numerical Python dealing with arrays and lists
import seaborn as sns  #for ploting graphs
import matplotlib.pyplot as plt #creating changes and lines in plots
import matplotlib.mlab as mlab
import matplotlib
```

```python
df = pd.read_csv(r'C:\Users\Shira\OneDrive\Desktop\movies.csv') 
df
```

```python
#true there are missing values, false there are not
df.isnull()
```

```python
df=df.drop_duplicates()
df
```

```python
df.sort_values(by=['gross'], inplace=False, ascending=False)
df
```

```python
#Plot data and a linear regression model fit
#asumption- higher budget => higher gross
sns.regplot(x="gross", y="budget", data=df, scatter_kws={"color": "red"}, line_kws={"color":"blue"})
plt.title('gross vs budget')

#asumption is true
```

```python
plt.scatter(x=df['gross'], y=df['budget'])
plt.title('gross vs budget')
```

```python
# Correlation Matrix between all numeric columns

df.corr(method ='pearson')
```

```python
correlation_matrix = df.corr()

sns.heatmap(correlation_matrix, annot = True)

plt.title("Correlation matrix for Numeric Features")

plt.xlabel("Movie features")

plt.ylabel("Movie features")

plt.show()
```

```python
#correlation in another form
corr_pairs = correlation_matrix.unstack()
print(corr_pairs)
```

```python
#sorting desc
sorted_pairs = corr_pairs.sort_values()
print(sorted_pairs)
```

```python
df.groupby(['company', 'year'])[["gross"]].sum()
```

```python
#changing object type to category in order to do corr calcu
df_numerized = df


for col_name in df_numerized.columns:
    if(df_numerized[col_name].dtype == 'object'):
        df_numerized[col_name]= df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes
        
df_numerized
```

```python
correlation_matrix = df_numerized.corr(method='pearson')

sns.heatmap(correlation_matrix, annot = True)

plt.title("Correlation matrix for Movies")

plt.xlabel("Movie features")

plt.ylabel("Movie features")

plt.gcf().set_size_inches(15, 15)




plt.show()
```
